```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = r'C:\Users\NIKHITHA\Downloads\Dataset.csv'
df = pd.read_csv(url, encoding='utf-8')



# Data Preprocessing
# Replace incorrect data in the "height" column with random numbers between 150 and 180.
df['height'] = np.random.uniform(150, 180, size=len(df))

# Handle missing or inconsistent data
# For example, if there are missing values in the "Age" column, you can fill them with the mean age.
df['Age'].fillna(df['Age'].mean(), inplace=True)

#1Count of Employees in Each Team
team_counts = df['Team'].value_counts()

# Display the number of employees in each team
print("Number of employees in each team:")
print(team_counts)

# Visualize the count of employees in each team
plt.figure(figsize=(12, 6))
sns.countplot(x='Team', data=df)
plt.title('Number of Employees in Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.show()

# 2. Position Segregation
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.countplot(x='Position', data=df,hue='Position', palette="magma", legend=False)
plt.title('Distribution of Employees Across Positions')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()

# 3. Age Group Analysis
# Visualize the distribution of ages
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=5, kde=False, color='skyblue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Number of Employees')

# Display the most common age group
most_common_age_group = df['Age'].mode().values[0]
plt.axvline(most_common_age_group, color='red', linestyle='dashed', linewidth=2, label=f'Most Common Age Group: {most_common_age_group}')
plt.legend()
plt.show()

# 4. High Spending Analysis
high_spending_team_position = df.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print(f'Team and Position with High Salary Spending: {high_spending_team_position}')
team_position_salary = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
team_position_salary = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

# Visualize the salary spending for each team and position
sns.barplot(x='Salary', y='Team', hue='Position', data=team_position_salary, palette='viridis')
plt.title('Salary Spending by Team and Position')
plt.xlabel('Total Salary Spending')
plt.ylabel('Team')
plt.show()


# 5. Correlation Analysis
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
sns.scatterplot(x='Age', y='Salary', data=df, hue='Team', palette='Set2')
plt.title('Correlation between Age and Salary')
plt.xlabelteam_position_salary = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()


```
